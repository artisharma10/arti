```python
data={'Name':['Alice','Bob','Charlie','David'],
      'Age':[25, np.nan,30, np.nan],
      'salary':[50000,60000,None,45000]}

df=pd.DataFrame(data)
print("Befor Cleaning:\n,df")


```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from math import isqrt
data=[10,12,13,12,11,200,14,13,15,12]
sns.boxplot(data)
plt.show()
Q1,Q3=np.percentile(data,[25,75])
IQR=Q3-Q1
lower,upper=Q1-1.5*IQR,Q3+1.3*IQR
outliers=[x for x in data if x <lower or x>upper]
print("outliers detected:",outliers)

```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pandas as pd
```

```python
df=pd.DataFrame({'Age':[18,22,30,40],
                  'salary':[20000,50000,80000,120000]})
scaler=MinMaxScaler()
df_norm=scaler.fit_transform(df)
scaler_std=StandardScaler()
df_std=scaler_std.fit_transform(df)
print("Original Data:\n",df)
print("\nNormalized:\n",df_norm)
print("\nStandardized:\n",df_std)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.DataFrame({
       'Name':['Alice','Bob','charlie'],
       'Age':[25,32,47],
       'Country':['US','India','US'],
       'Spend1':[200,150,300],
       'Spend2':[100,200,400]
})
df['Total_Spend']=pd.cut(df['Age'],bins=[0,30,50],labels=['Young','Middle'])
encoder=LabelEncoder()
df['Country_encoded']=encoder.fit_transform(df['Country'])
print(df)



```
